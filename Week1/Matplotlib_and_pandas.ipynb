Exercise 1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: â€“

Line Style dotted and Line-color should be red  
Show legend at the lower right location.  
X label name = Month Number  
Y label name = Profit in dollar  
Add a circle marker.  
Line marker color as read  
Line width should be 3  

```python
#code here
import matplotlib.pyplot as plt
from google.colab import drive
import pandas as pd
drive.mount('/content/gdrive')
path='/content/gdrive/My Drive/Colab Notebooks/company_sales_data.csv'
read=pd.read_csv(path)
plt.plot(read['month_number'], read['total_profit'],linestyle='dotted', color='red', marker='o', markersize=8, linewidth=3, label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollars')
plt.title('Total Profit Over Months')
plt.legend(loc='lower right')
plt.show()
```

Exercise 2: Read all product sales data and show it  using a multiline plot.
 Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline and different color for each product ).

```python
#code here
import matplotlib.pyplot as plt
from google.colab import drive
import pandas as pd
drive.mount('/content/gdrive')
path='/content/gdrive/My Drive/Colab Notebooks/company_sales_data.csv'
read=pd.read_csv(path)
plt.plot(read['month_number'], read['facecream'], color='red', label='facecream')
plt.plot(read['month_number'], read['facewash'], color='pink', label='facewash')
plt.plot(read['month_number'], read['toothpaste'], color='blue', label='toothpaste')
plt.plot(read['month_number'], read['bathingsoap'], color='black', label='bathingsoap')
plt.plot(read['month_number'], read['moisturizer'], color='orange', label='moisturizer')
plt.plot(read['month_number'], read['shampoo'], color='purple', label='shampoo')
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('units wrt products')
plt.title('month wise unit sale for products')
plt.show()
```

Exercise 3: Calculate total sale data for last year for each product and show it using a Pie chart  
Note: In Pie chart display Number of units sold per year for each product in percentage.


```python
#code here
import matplotlib.pyplot as plt
from google.colab import drive
import numpy as np
import pandas as pd
drive.mount('/content/gdrive')
path='/content/gdrive/My Drive/Colab Notebooks/company_sales_data.csv'
read=pd.read_csv(path)
products=np.array(['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'moisturizer', 'shampoo'])
total=read['total_units'].sum()
def sum_units(a):
  result=np.zeros_like(a)
  for i in range(len(a)):
    p=read[a[i]].sum()
    result[i]=(p/total)*100
  return result
result_arr=sum_units(products)
plt.pie(result_arr, labels=products,autopct='%1.1f%%', startangle=90)
plt.title('units sold wrt products')
plt.show()



```
