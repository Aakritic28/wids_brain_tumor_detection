In this part of the assignment you have to implement multivariate gradient descent to find the minimas (local and global) of the given function:
Note : you can find different minimas by changing your initialisation.  
$f(x) = x^4 + x^2y^2 - y^2 + y^4 + 6$

```python
#start your code here :)
import numpy as np
import matplotlib.pyplot as plt
def f(x,y):
  return x**4 + x**2 * y**2 - y**2 + y**4 + 6
def df_dx(x,y):
  return 4*x**3 + 2*x *y**2
def df_dy(x,y):
  return 2*y *x**2 - 2*y +4*y**3
def gradient_descent(x,y, iterations, rate):
  for i in range(iterations):
    x=x-rate*df_dx(x,y)
    y=y-rate*df_dy(x,y)
    print(f"After {i+1} iteration: x = {x}, y = {y}, f(x,y) = {f(x,y)}")
  return x, y, f(x,y)

iterations=100
rate=0.01
initial_x=1.0
initial_y=1.0

final_x, final_y, min_value= gradient_descent(initial_x, initial_y, iterations, rate)
print(f"\n Minima found at : x= {final_x}, y={final_y}, minimum value of f(x,y) ={min_value}")
```
